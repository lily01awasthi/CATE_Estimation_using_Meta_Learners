# CATE Estimation and Evaluation

```python
# CATE_estimation_evaluation.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path to the data with CATE estimates
s_cate_estimates_path = '../results/s_predictions.csv'
t_cate_estimates_path = '../results/t_predictions.csv'
x_cate_estimates_path = '../results/x_predictions.csv'
r_cate_estimates_path = '../results/r_predictions.csv'
```

```python
#Load data with cate estimates
s_data_with_cate = pd.read_csv(s_cate_estimates_path,nrows=10)
t_data_with_cate = pd.read_csv(t_cate_estimates_path,nrows=20)
x_data_with_cate = pd.read_csv(x_cate_estimates_path,nrows=30)
r_data_with_cate = pd.read_csv(r_cate_estimates_path,nrows=10)
```

```python
import numpy as np
data = pd.DataFrame({
    'Index': np.arange(1, 11),
    'S-Learner': s_data_with_cate['s_CATE'],
    'T-Learner': t_data_with_cate['t_CATE'],
    'X-Learner': x_data_with_cate['x_CATE'],
    'R-Learner': r_data_with_cate['r_CATE']
})
# Melting the DataFrame to long format for seaborn
data_melted = data.melt(id_vars='Index', value_vars=[ 'S-Learner','T-Learner','X-Learner','R-Learner' ],
                        var_name='Meta-Learner', value_name='CATE Estimate')

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=data_melted, x='Index', y='CATE Estimate', hue='Meta-Learner', marker='o')
plt.title('Comparison of CATE Estimates by Different Meta-Learners')
plt.xlabel('Index')
plt.ylabel('CATE Estimate')
plt.legend(title='Meta-Learner')
plt.grid(True)
plt.show()
print('X-Learner and T-Learner are exactly overlapping')
```
